```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line',marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
customer_df = df.groupby('Customer_ID').agg({
    'Revenue':'sum',
    'Order_ID':'count'}).rename(columns={
    'Revenue':'Total_Spend',
    'Order_ID':'Total_Orders'})

customer_df['Avg_Order_Value']=customer_df['Total_Spend'] / customer_df['Total_Orders']

customer_df.info()
customer_df.head()
```

```python
features =customer_df[['Total_Spend','Total_Orders','Avg_Order_Value']]
features.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans as km

kmeans=km(n_clusters=3,random_state=42)
customer_df['Segment']=kmeans.fit_predict(scaled_features)

customer_df.head()
```

```python
customer_df['Segment'].value_counts()
```

```python
customer_df['Segment']
```

```python
segment_summary = customer_df.groupby('Segment')[['Total_Spend','Total_Orders','Avg_Order_Value']].mean()
segment_summary
```

```python
import matplotlib.pyplot as plt

# Map segments to readable names
segment_labels = {
    2: "High Value Customers",
    0: "Medium Value Customers",
    1: "Low Value Customers"
}

# Colors for each segment
colors = {
    2: "red",
    0: "blue",
    1: "green"
}

plt.figure(figsize=(10, 6))

# Plot each segment separately
for segment in customer_df['Segment'].unique():
    seg_data = customer_df[customer_df['Segment'] == segment]
    plt.scatter(
        seg_data['Total_Orders'],
        seg_data['Total_Spend'],
        label=segment_labels[segment],
        color=colors[segment],
        alpha=0.7
    )

plt.xlabel("Total Orders")
plt.ylabel("Total Spend")
plt.title("Customer Segmentation (K-Means Clusters)")
plt.grid(True)  # Grid on

plt.legend(title="Customer Segments")  # Legend added
plt.show()
```
